```python
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
df =sns.load_dataset('titanic')
```

```python
df.head()
```

```python
#https://www.kaggle.com/competitions/titanic/data   // we can download csv file

#servived : outcome of servived(0 = No, 1 = Yes)
#Pclass : socio-economic class(1 = Upper class, 2 = middle class, 3 = lower class)
#sibSp:  number of sibling and spouse of the passenger abord
# Parch : No. of parents and children of the passenger abord
# Embarked : part of embarkation of the passenger (c = cherbourg, Q = queenStown, S = southampton)
```

Scatter Plot

```python
#palette = ["red", "green", "blue"], #hue="class" 
#size = "pclass"

sns.scatterplot(x = "age", y = "fare", data = df, palette = ["red", "green", "blue"], hue = "class", size = "pclass") 
```

Histogram & Distribution plot

```python
 #kde = True , it show order of the line for smoothness curve
 #bins = 15  // it devide pno. of parts
 # histogram and distplot both are same 

sns.histplot(df["age"], kde = True, bins = 15)
```

```python
sns.displot(df["age"], kde = True, bins = 15)
```

Barplot

```python
sns.barplot(x = df["sex"], y = df["fare"], palette = ["red", "green"])
```

Box Plot
# Box plot will find out outliner from graph

```python
sns.boxplot(x = df["pclass"], y =df["fare"], palette=["red", "green"], hue = df["sex"])  # hue = df["sex"]
```

Violin plot


```python
sns.violinplot(x = df["pclass"], y = df["fare"], hue= df["sex"], palette = ["red", "green"])
```

Strip Plot

#dodge = True   //it devide male and female fields
# it how actial plot value

```python
sns.stripplot(x = df["pclass"], y = df["fare"], hue = df["sex"], palette=["red","green"], dodge = True)
```
